## **<span style="color:black">9. 딥러닝 모델링 & 모델 성능평가 및 그래프 출력</span>**

<br>
<font color=blue>9-1. 아래 가이드대로 해지여부를 분류하는 딥러닝 모델을 만드세요.</font>

+ 첫번째 Hidden Layer : unit 64 , activation='relu'
+ 두번째 Hidden Layer : unit 32 , activation='relu'
+ 세번째 Hidden Layer : unit 16 , activation='relu'
+ 각 Hidden Layer 마다 Dropout 0.2 비율로 되도록 하세요.
+ EarlyStopping 콜백을 적용하고 ModelCheckpoint 콜백으로 validation performance가 좋은 모델을 h5 모델로 저장하세요.
+ batch_size는 10, epochs는 10으로 설정하세요.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Input, Model
from keras.optimizers import Adam
from keras.models import Sequential
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import Embedding, BatchNormalization, Dense, Flatten, Dropout, Input, Conv1D, GlobalMaxPooling1D, LSTM, MaxPooling1D
from sklearn.metrics import accuracy_score
```

```python
# 여기에 답안코드를 작성하세요.
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(train_X.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('binary_classification_model.h5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')

# 모델 훈련
history = model.fit(train_X, train_y, epochs=10, batch_size=10, validation_split=0.2, callbacks=[early_stopping, checkpoint])
```

```python
y_pred = model.predict(test_X) # 딥러닝 모델이 sigmoid() 함수를 사용하기 떄문에 각각 1개 관측치가 0 ~ 1 값의 범위를 가진다.
y_pred
```

```python
y_pred = np.argmax(y_pred, axis=1) # 기준을 마련해서 어느정도에 도달하면 0으로 최종 예측하고 아니면 1로 최종 예측한다.
y_pred
```

```python
print('정확도 : ', accuracy_score(test_y, y_pred))
```

```python
# 여기에 답안코드를 작성하세요.
train_y_onehot = tf.keras.utils.to_categorical(train_y)
test_y_oneshot = tf.keras.utils.to_categorical(test_y)

model2 = Sequential()

model2.add(Dense(64, activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(32, activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(16, activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(2, activation='softmax'))

model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
checkpoint = ModelCheckpoint('binary_classification_model.h5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')

# 모델 훈련
model2.fit(train_X, train_y_onehot, epochs=10, batch_size=10, callbacks=[early_stopping, checkpoint])
```
